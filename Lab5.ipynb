Given a $\mathbb{R}=1$ vector in a 2-dimensional plane and visualize it using the cells below. In the first markdown cell below, write your vector using LaTeX. You should represent it as a system of linear equations. In your methodology discuss the different functions that you have used, the equation of your vectors, and the flowchart for declaring and displaying your vector. $
A = \begin{bmatrix} 
1 \ 1 \ 0 \ 2 \\
-1 \ -1\ 0 \ -2 \\
\end{bmatrix}$ 

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


A = np.array([
    [1, 1, 0, 2],
    [-1, -1, 0, -2]
])

plt.xlim(0,2)
plt.ylim(0,2)



plt.quiver([0,0],[0,0], A[:,0], A[:,1], 
           angles='xy', scale_units='xy',scale=1, 
           color=['red','blue'])
plt.show()
```

$$A = \begin{bmatrix} 
1 \ 1 \ 0 \ 2 \\
-1 \ -1\ 2 \ 0 \\
\end{bmatrix} $$

```
A = np.array([
    [1, 1, 0, 2],
    [-1, -1, 2, 0]
])
fig = plt.figure()
ax1 = fig.gca(projection='3d')

ax1.set_xlim([0, 5])
ax1.set_ylim([0, 5])
ax1.set_zlim([0, 5])

origin = (0,0,0)
ax1.quiver( origin, origin, origin, A[0,0], A[0,1], A[0,2], 
           arrow_length_ratio=0.05,linestyles= '-',colors=['red'])
ax1.quiver( origin, origin, origin, A[1,1], A[1,2], A[1,3], 
           arrow_length_ratio=0.05,linestyles= '-',colors=['blue'])

plt.show()
```
